```python
# Importing the data

import pandas as pd
import numpy as np
import os
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model.logistic import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
```

```python
#Checking files in current directory
os.listdir()
```

```python
# Creating Dataframe
df = pd.read_csv('smsspamcollection/SMSSpamCollection', delimiter = '\t', header = None)
df.head()
```

```python
#slitting data set  into training and testing set
x_train, x_test, y_train, y_test = train_test_split(df[1], df[0])
```

```python
# Creating TFIDF Vectorizer
vectorizer = TfidfVectorizer()

#Transforming Training set
x_train_f = vectorizer.fit_transform(x_train)
x_test_f = vectorizer.transform(x_test)
```

```python
# Visualizing the values 
print(x_train[0])
print(x_train_f[0])
print(y_train[0])
```

```python
# Creating and Fitting the model
clf = LogisticRegression()
clf.fit(x_train_f,y_train)
```

```python
#pridicting the model
for i in range(len(x_test)):
    #print(x_test.iloc[i], clf.predict(x_test_f[i]))
    print("S.No. "+str(i) + ": Message : {0} -- Prediction : {1}\n".format(x_test.iloc[i], clf.predict(x_test_f[i])))
```

```python
#Prediction for calculation of Performance matrix
y_test_pred = clf.predict(x_test_f)
```

```python
y_test_pred = pd.DataFrame(y_test_pred)
y_test_pred[0] = y_test_pred[0].astype('category')
ytest = np.array(y_test_pred[0].cat.codes).astype(int)


y_test = pd.DataFrame(y_test)
y_test[0] = y_test[0].astype('category')
y_test = np.array(y_test[0].cat.codes).astype(int)

print(ytest)
```

```python
cm = confusion_matrix(y_test, ytest)
plt.matshow(cm)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
print(cm)
```

```python
print ('Accuracy:', round(accuracy_score(y_test, ytest)*100,2),'%')
```
