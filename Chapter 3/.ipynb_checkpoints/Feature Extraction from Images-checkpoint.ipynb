```python
import numpy as np
from skimage.feature import corner_harris, corner_peaks
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import skimage.io as io
from skimage.exposure import equalize_hist
```

```python
def show_corners(corners, image):
    fig = plt.figure()
    plt.gray()
    plt.imshow(image)
    y_corner, x_corner = zip(*corners)
    plt.plot(x_corner, y_corner, 'or')
    plt.xlim(0, image.shape[1])
    plt.ylim(image.shape[0], 0)
    fig.set_size_inches(np.array(fig.get_size_inches()) * 1.5)
    plt.show()
```

```python
mandrill = io.imread('/root/Mastering Machine Learning With Scikit-Learn By Gavin Hackeling/Chapter 3/Images/Test Image 1.png')
mandrill = equalize_hist(rgb2gray(mandrill))
# WIth Variation in min_distances

corners = corner_peaks(corner_harris(mandrill), min_distance=20)
show_corners(corners, mandrill)

corners = corner_peaks(corner_harris(mandrill), min_distance=10)
show_corners(corners, mandrill)

corners = corner_peaks(corner_harris(mandrill), min_distance=5)
show_corners(corners, mandrill)
```
